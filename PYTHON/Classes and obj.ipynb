```python
### defining a class with instance method
class Dog:
    def __init__(self,name,age): 
        self.name=name
        self.age=age
    def bark(self) :
        print(f"{self.name} says woof")  


dog1=Dog("RESS",3) 
dog1.bark() 
dog2=Dog("lucy",4)   
dog2.bark()   
dog3=Dog("KAIN",2)   
dog3.bark()
         
```

```python
#### Modelling a Bank account
###  Define a class for bank account

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print (f"{amount} is deposited.New balance is {self.balance}")
    
    def withdrawn(self,amount):
        if amount>self.balance:
            print("insufficient funds")
        else:
            self.balance =-amount    
            print(f"{amount} is withdrawn,new balance is {self.balance}")
    def get_balance(self):
        return self.balance
    
   
 
```

```python
#### CREATE ACCOUNT
account = BankAccount("krish",5000)
print(account.balance)

### call instance methods
account.deposit(100)
account.withdrawn(300)
    
    
print(account.get_balance())
```

INHERITENCE IN PYTHON:This is a concept in OOP that allows a clas to inherit attributes and methods from another class

```python
### Inheritance (single inheritance)
### Parent class
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
     
    def drive(self):
        print(f"Colin will drive the {self.enginetype} car ")
        
     
```

```python
car1=Car(4,5,"diesel")
print(car1.drive())
```

```python
### super()that is calling the parent class

class Honda (Car):
    def __init__(self,doors,windows,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    def sefdriving(self):
       print(f"Honda supports self driving :{self.is_selfdriving}")
       
Honda1=Honda(4,5,"electric",True)   
Honda1.sefdriving()   
```

```python
### multiple inheritance --when a class inherit more than more base class.
### base class1
class Animal:
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print("subclass must implement this method") 
        
   ### base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner
    ### Derived class
class Dog(Animal,Pet):
        def __init__(self,name,owner):
            Animal.__init__(self,name)
            Pet.__init__(self,owner)
        def speak(self):
            return f"{self.name } say woof"
  ### creating an object
dog = Dog("buddy","colin")  
print(dog.speak())
print (f"ownwer:{dog.owner}")
        
```

POLYMOPHISIM allows objects of different classes to be treated as object of a common superclass,
it also provide a way to perform single action in different forms

```python
### Method Overrding -- this allows a child class to provide a specific method that is already defined in its parent class
#### BASE CLASS
class Animal:
    def speak(self):
        return "sound of the animal"
### Derived class 1
class Dog(Animal):
    def speak(self):
        return "woof"
### Derived class
class Cat(Animal):
    def speak(self):
        return "meow"
    
### Function that demonstrates polymorphism
def animal_speak(animal):
    print(animal.speak())

   
        
dog=Dog()
cat=Cat()
print (dog.speak())
print (cat.speak())
animal_speak(dog)
animal_speak(cat)
```

```python
class Person:
    def __init__ (self,name,age,gender):
        self._name= name   ### protected variables
        self._age = age   ### protected var
        self.gender = gender
class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

employee = Employee("colin",22,"male")
print(employee._name) 
```
