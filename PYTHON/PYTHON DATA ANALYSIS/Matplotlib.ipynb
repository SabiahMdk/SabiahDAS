```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]

### CREATE a line plot
plt.plot(x,y)
plt.xlabel('x axix')
plt.ylabel('y axix')
plt.title("Basic Line Plot")
plt.show()
```

```python
x= [1,2,3,4,5,6]
y = [11,19,8,17,13,15]

### Create a customize line plot
plt.plot(x,y ,color = 'red',linestyle='-.',marker ='o')
### linewidth-- to increase the line thickness
## markersize --- for the thickness of the marker
```

```python
### BAR PLOT
Categories = ['a','b','c','d','e','f']
values = [3,6,7,5,4,8]
## Create a bar plot
plt.bar(Categories,values,color = 'green')
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar Plot')
```

```python
####  Histogram ----used to represent the distribution of a dataset
```

```python
## sample data
data = [1,2,2,2,3,3,4,4,5,5]
### Create Histogram
plt.hist(data,color='green',edgecolor='BLACK')
plt.xlabel('data')
plt.ylabel('values')
plt.title('Histogram Plot')
```

```python
### create a scatter plot
### Sample data
x = [1,2,4,3,5,6]
y= [2,4,7,2,3,5]

plt.scatter(x,y,color ='red',marker = 'x')
```

```python
### PIECHARTS
```

```python
labels=['A','B','C','D']
sizes = [20,30,35,10]
colors=['gold','green','blue','red']
# explode=(0.2,0,0,0)  to move out the 1st slide
#### creating pie chart
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=3)

```
