```python
### SERIES --- a one dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, python objects, etc.). A pandas Series is like a column in a table. It is a one-dimensional array holding data of any type.
```

```python
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("series \n",series)
print(type(series))
```

```python
### create series from dictionary
data = {'a':3,'b':2, 'c': 4, 'd':5}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30,67,34] 
index =['a','b','c','d','e']
pd.Series(data,index)
        
```

```python
### Data frames --- a table that is multidimension: having multiple rows and columns
### create a dataframe from a dictionary 
data = {
    'name':['john','jack','colin'],
    'age':[22,34,23],
    'city':['kenya','india','egpyt']
}
df=pd.DataFrame(data)
print(df)
print(type(data))
print (type(df))
```

```python
import numpy as np
np.array(df)
```
