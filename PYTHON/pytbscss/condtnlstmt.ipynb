```python
### calculator
## take user input
num1 = int(input("enter your first number"))
num2 = int(input("enter your second number"))
operator =input("enter operators(+, - , / , *)")

### perform the operation
if operator == "+":
    result = num1 + num2 
elif operator == "-":
     result = num1 - num2 
elif operator == "/":
     result = num1 / num2 
elif operator == "*":
     result = num1 * num2          
else:
        print ("invalid operator")
print(result)    
```
