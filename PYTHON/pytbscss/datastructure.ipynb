```python
### accessing list elements
## we are indexing
fruits = ["apple","banana","cherry","orange"]
print (fruits)
print(fruits [0])
print(fruits[3])
print (fruits[1])
```

```python
 print (fruits[0:])
 print (fruits[1:])
 print (fruits[1:3])
```

```python
## modifying list elements --- changing or replacing any element in the list 
fruits = ["apple","banana","cherry","orange"]
print (fruits)
fruits[3] = "watermelon"
print (fruits)
fruits [2] = "mango"
print (fruits)
```

```python
## list method
###append----this will simply add an item to the end
### insert ---- this will add at the middle or select any index
fruits = ["apple","banana","cherry","orange"]

fruits.append("cherry")
print(fruits)

fruits.insert(3,"pineapple ")
print(fruits)

fruits.insert(1,"guava")
print(fruits)

fruits.remove("apple")
print(fruits)


#### remove and return the last elements
popped_fruits=fruits.pop()
print (popped_fruits)
print(fruits)

index=fruits.index("orange")
print(index)

fruits.insert(3,"cherry")
print(fruits.count("cherry"))

fruits.sort #--- sort in ascending order
fruits.clear#---- to clear all list
```

```python
#### slicing list
number = [1,2,3,4,5,6,7,8]
print(number[2:5])
print (number[:5])
print (number[1:])
print (number[5:
])
print (number[::2])
print(number[::3])
print(number[::-2])
```

```python
### list comprehension
list = []
for x in range(10):
   list.append(x**2)
   print (list)
```

```python
[x**2 for x in range(10)]##### basic syntax list comprehension
```

```python
even_numbers= [num for num in range(10) if num%2==0] ###---list comprehension with condition
print(even_numbers)
```

```python
### list comprehension with function calls

words= ["mark", "abstract","python","carol"]
lenght = [len(words) for words in words]
print(lenght)
```
