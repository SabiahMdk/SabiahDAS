```python
#### modifying dictionary element
###---- dictionary are immutable,you can add,change or update 
```

```python
employee = {"name": "Mark","dept":"compliance","position":"support"}
print (employee)
employee ["dept"] = "IT"  ### updating the key "dept"
print(employee)
employee ['adress'] = 'maryland' ##adding a new key "adress "
print(employee)
```

```python
del employee ['position'] ### deleting the key 'position'
print (employee)
```

```python
### dictionary methods
keys = employee.keys()### getting all the keys
print (keys)
values = employee.values()
print(values)
items = employee.items()
print(items)
```

```python
### iterating over dictionaries ### the use of loops to itearte over keys,values or items
for keys in employee.keys():
    print (keys)
```

```python
### iterate over values
for values in employee.values():
    print (values)
```

```python
#### iterate over key value pairs
for key,value in employee.items():
    print(key,value)
    print (f"{key} :{value}")### i do not run this
```

```python
#### NESTED DICTIONARIES --- A DICTIONARY inside another dictionaries
employees = {
       "employee1":{"name":"steven", "dept":"IT"},
       "employee2":{"name":"mark", "dept":"HR"},
}

print(employees) ### iterating
for keys in employees.keys():
    print (keys)

### access nested dictionaries elements
print(employees["employee2"] ["name"])
print(employees["employee2"] ["dept"])
```

```python
#### iterating nested dictionaries 
employees.items()
```

```python
for employee_id,employee_info in employees.items():
 print (employee_id,employee_info)
 print( f"{employee_id}:{employee_info}")
for key,value in employee_info.items():
 print(f"{key}:{value}")
```

```python
### DICTIONARY COMPREHENSION
squares ={x:x**2 for x in range(5)}
print(squares)
```

```python
### use a dictionary to count the frequency of elememts in list
numbers = [1,2,3,3,2,3,7,7,8,4,5,6,7,8]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency [number]+=1
    else:
        frequency [number] = 1
        print(frequency) 
        
```

```python
 #### MERGE 2 DICTIONARIES INTO ONE
dict1= {"a":1, "b":2}
dict2 = {"b" :3,"c" :4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
