This is to read and write file both text and binary


```python
### Read a whole file
with open('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
### read a file line by line
with open ('example.txt','r') as file:
    for line in file:
        print(line)
        
      #### strip() removes the newline character print(line.strp())
```

```python
#### writing a file (overwriting)
### w = writing method, which enables you to write
with open('example.text','w') as file:
 file.write('hello world!\n')
 file.write('this is a new line.')
 
```

```python
### write a file (without overwriting)
with open ('example.txt','a') as file:
    file .write ("append operation taking")
```

```python
### writing a list of lines to a file
lines = ['\n first line\n','second line\n', 'third line\n']
with open ('example.txt','a') as file:
    file.writelines(lines)
    
```
