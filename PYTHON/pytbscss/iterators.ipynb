Iterators

These are advanced python concepts that allow efficient looping and memory management, they provide a way to access elements of a collection sequentially.

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
  
```

```python
print(my_list)
```

```python
### iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
### iterate through all the element
### the next element in iterator print the element 1 ,2,3 as it is displayed as next

next(iterator)
```

```python
random_list = [3,5,6,7]
print(random_list)
```

```python
iterator = iter(random_list)
next(iterator)
```

```python
try:
   print(next(iterator))
except StopIteration:
    print("there are no elements in the iteration")   
    ### it displays all the random_list and at the end the nessage
```

```python
### string iterator

my_string = "hello"
string_iterator = iter(my_string)
print(next(string_iterator))
print(next(string_iterator))
```

GENERATORS 


Generators are a simpler way to create iterators, they produce a series of value lazily and store them in memory


```python
def square (n):
    for i in range(3):
        yield 1 ** 2
```

```python
square (3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(a) 
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
    
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

```python
### practical: reading large files ---generators are useful for reading large files which allow you to process one line at a time without loading it entirely 

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'large_file.txt'
for line in read_large_file(file_path):
    print(line.strip())
```
