```python
#### creating a to do list to keep track of task
to_do_list = ["clean the house","wash my clothes","buy groceries"]
print (to_do_list)

### adding to task
to_do_list.append("pay bills")
to_do_list.append("play a home game")
print (to_do_list)

### removing a completed task
to_do_list.remove("clean the house")
print (to_do_list)

### checking if a task is in the list
if "pay bills" in to_do_list:
    print("dont forget to pay the utility bills")
print("to do list remaining")
for task in to_do_list:
    print(f"-{task}")

```

```python
#### ORGANISING STUDENT GRADES
### Create a list to store and calculate average grades for student

#### organizing student grades
grades= [54,70,45,88,30,43]

#### adding a new grade
grades.append(67)
print(grades)

### calculating the average grade
average_grade = sum(grades) / len(grades)
print(f"average grade:{average_grade:.2f}")

### finding the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"highest grade:{highest_grade}")
print(f"lowest grade:{lowest_grade}")
```

```python
##### Managing an inventory
### use a list to manage inventory items in a store 

inventory = ["sugar", "rice", "pasta","flour"]
inventory.append("edible oils")
print(inventory)
 
### removing an item from the list
inventory.remove("rice")
print(inventory)

### checking if an item is in the stock or not
item = "salt"
if item in inventory:
    print(f"{item} is in stock")
else:
    print(f"{item} is out of stock")  

#### printing the inventory
print("inventory list:")  
for item in inventory:
    print(f"-{item}")
```
