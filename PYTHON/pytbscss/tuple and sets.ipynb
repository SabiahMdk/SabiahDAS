```python
#### SETS----they are used to store collection of unique items, they are used for membership tests, eliminating duplicate entries and performing mathematical set operations like union, intersection and difference.

```

```python
####  create a set {}

my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_set = set([1,2,3,4,5,6,3,2,3,4,1]) #### it eliminate duplicate entries
print(my_set)
```

```python
### basic sets operations
## adding and removing elements
my_set.add(7)
print(my_set)
my_set.add(9)
print(my_set)
```

```python
#### i had already removed 6 and 3 and run it

my_set.remove(2)
print(my_set)
```

```python
### pop method

removed_element = my_set.pop()
print(removed_element)
print(my_set) 
```

```python
#### Clear all the elememts
my_set.clear()
print(my_set)
```

```python
### set membership set
my_set = {1,2,3,4,5}
print(3 in my_set)
print (10 in my_set)
```
