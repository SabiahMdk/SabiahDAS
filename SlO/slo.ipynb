```python
### Data serialization
import json
data = {'name':'colin','age': 23}
json_str=json.dumps(data)
print(json_str)
print (type(json_str))


parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
### csv
import csv

with open('example.csv',mode='w',newline='')as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['charles',27])
    
    with open ('example.csv', mode='r')as file:
        reader=csv.reader(file)
        for row in reader:
            print(row)
            
```

```python
### datetime

from datetime import datetime,timedelta
now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
### regular expression ---to match a specific in a data set

import re
pattern =r'\d+'
text = 'there are 123 workers'
match = re.search(pattern,text)
print(match.group())
```
